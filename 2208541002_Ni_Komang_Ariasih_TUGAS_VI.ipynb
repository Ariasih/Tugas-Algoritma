***5.21.7 Pengelompokan data menggunakan Pandas***

Data yang sama dapat dikelompokkan untuk analisis lebih lanju

```python
#menggunakan fungsi metode.groupby
#menggunakan fungsi perintah .mean
bunga_iris_csv.groupby(["sepal length (cm)"]).mean()
#metode 'groupby()'untuk mengelompokkan baris di Data Frame berdasarkan nilai dari kolom 'sepal width(cm)'
#'mean()' untuk menghitung nilai rata-rata dari setiap kelompok
#Ketika kode diatas dijalankan, akan dihasilkan suatu data bru yang berisi nilai rata-rata dari setiap kolom 'bunga iris' yg sudah dikelompokkan berdasarkan nilai kolom 'sepal width (cm)'
```

```python
bunga_iris_csv.groupby(["sepal length (cm)"]).mean().sort_values("petal width (cm)",ascending=True)
#metode 'groupby()' untuk mengelompokkan baris dalam DataFrame berdasaekan nilai 'sepal length(cm)'
#metode 'mean()' untuk menghitung rata-rata dari setiap kelompok.
#metode 'sort_values()' untuk mengurutkan data hasil kelompokkan berdasarkan kolom 'petal width (cm)' secara menaik
#ketika kode tersebut dijalankan, output yang dihasilkan adalah data baru yang berisi 
#nilai rata-rata dar setiap kolom yang sudah dikelompokkan berdasarkan nilai pada kolom 'sepal length (cm)
#Kemudian, data diurutkan berdasarkan kolom 'petal width (cm)' secara menaik
```

```python
#perintah sum()
bunga_iris_csv.groupby(["sepal width (cm)"]).sum()
#pada kode diatas dilakukan suatu pengelompokkan data pada variabel bunga_iris berdasarkan kolom 'sepal width (cm)'
#Selanjutnya dilakukan agregasi nilai dengan fungsi 'sum()' 
#dimana untuk setiap kelompok data yang dihasilkan dari pengelompokkan tersebut
#Output yang dihasilkan adalah suatu data hasil penjumlahan untuk setiap kolom dari data bunga_iris 
#dengan perhitungannya berdasarkan kelompok-kelompok yang dibentuk berdasarkan nilai unik pada kolom 'sepal width(cm)'
```

```python
#perintah .count()
bunga_iris_csv.groupby(["petal width (cm)"]).count()
#perintah grupby() digunakan untuk mengelompokkan data berdasaekan nilai unik pada suatu kolom yaitu 'petal width (cm)'
#perintah count() digunakan untuk menghitung jumlah entri data pada setiap kelompok
#hasil yang dikeluarkan adalah jumlah entri data pada setiap nilai unik pada kolom 'petal width (Cm)'
```

***5.22.1 Grafik sederhana dengan Matplotlib***

```python
a = [1,2,3]
b = [2,4,6]
plt.plot (a,b)
#metode plot digunakan untuk menggambar garis yang menghubungkan ketiga titik yaitu (1,2), (2,4), dan (3,6) pada bidang koordinat
plt.title ("Grafik b = 2a")
#tittle digunakan untuk memberikan judul grafik yaitu grafik b = 2a
plt.xlabel ('a')
#kemudian, sumbu x diberi label dengan kode seperti diatas
plt.ylabel ('b')
#kemudian, sumbu y diberi label dengan kode seperti diatas
```

***5.22.2 Properti Pembuatan Grafik dengan Matplotlib***

```python
#diinisialisasikan variabel x dan y yang merupakan list yang berisi 9 nilai dari masing-masing variabel x dan y yang berkaitan dengan fungsi y = x^3
x = [-4,-3,-2,-1,0,1,2,3,4]
y = [16,9,4,1,0,1,4,9,16]
#Kemudian, diinisialisasikan juga variabel x1 dan y1 yang juga merupakan list yang berisi masing-masing anggota dari variabel x1 dan y1 
x1 = [1,2,3,4,5,6,7]
y1 = [-2,0,2,4,6,8,10]

#menentukan ukuran dan resolusi gambar
plt.figure(figsize=(2,2), dpi =300)
#figsize digunakan untuk menentukan gambar dengan format inci. sehingga pada kode ini ukuran gambar yang akan dihasilkan adalah 2x2 inci
#dpi digunakan untuk menentukan resolusi gambar dalam piksel per inci. Sehingga, pada kode ini resolusi yang dihasilkan adalah 300 piksel per inci

#Kemudian ditentukan font, ukuran font, serta warna dari garis dan grafik
plt.plot(x,y,label='y=x^2',color='blue',linewidth=3,marker='.',markersize =10,markeredgecolor='purple',linestyle='-')
#label='y=x^2' : dugunakan untuk memberi label atau judul pada plot
#color : digunakan untuk memberi warna pada plot
#linewidth : digunakan untuk memberi ketebalan garis pada plot
#marker : digunakan untuk menandai titik pada plot dengan bentuk titik
#markersize : digunakan untuk menentukan ukuran dari marker
#markerredgecolor : digunakan untuk memberi warna pada tepi marker
#linestyle : digunakan untuk menentukan jenis garis pada plot
plt.plot(x1,y1)
#metode ini digunakan untuk membuat plot dari data pada variabel x1 dan y1
#plot ini tidak memiliki variabel lain seperti label ataupun color sehingga plot ini akan menggunakan nilai default untuk setiap parameter
plt.title ('Grafik y =x^2', fontdict={'fontname' : 'Times New Roman', 'fontsize' :14})
#Ditentukan font dan ukuran font pada judul grafik
#'Grafik y =x^3', adalah string yang akan ditampilkan sebagai judul grafik
#fontdict : digunakan untuk mengatur jenis font dan ukuran font dari judul grafik
#fontname : jenis fnt
#fontsize : ukuran font
plt.xlabel ('x', fontdict={'fontname':'Comic Sans MS','fontsize':14})
plt.ylabel ('y', fontdict={'fontname':'Comic Sans MS','fontsize':14})
#Kemudian, sumbu x dan y diberi label pada grafik
plt.xticks ([-4,-3,-2,-1,0,1,2,3,4])
plt.yticks ([16,9,4,1,0,1,4,9,16])
#Digunakan fungsi plt.ticks untuk menentukan skala sumbu x dan t yang akan ditampilkan pada grafik
#pada sumbu x ditampilkan angka dari -4 sampai 4
#pada sumbu y ditampilkan angka dari 16 sampai 16
plt.legend()
#metode ini digunakan untuk menampilkan legenda pada grafik yang menunjukkan label garis untuk grafik pertaman
plt.savefig('grafik.png', dpi=300)
#metode ini digunakan untuk menyimpan grafik sebagai file gambar dalam format PNG dan nama Grafik "grafik.png" dengan resolusi 300dpi
plt.show()
#Metode ini digunakan untuk menampilkan grafik pada jendela hasil
```

***5.22.4 Properti Pembuatan Histogram dengan Matplotlib***

```python
#Properti pembuatan Histogram
#Diinisialisasikan variabel idol kpop BTS yang berisi list untuk menjadi label pada sumbu x
bangtan = ['RM', 'Jin','Suga', 'J Hope', 'Jimin', 'V', 'Jungkook']
#diinisialisasikan juga variabel tinggi sebagai nilai pada sumbu y
tinggi = [175, 173, 172, 173, 170, 173,173]

#digunakan method "plt.figure" untuk menentukan ukuran dan resolusi gambar
plt.figure(figsize=(10,5))
#figsize digunakan untuk menentukan gambar dengan format inci. sehingga pada kode ini ukuran gambar yang akan dihasilkan adalah 10x5 inci
hist = plt.bar(bangtan,tinggi,color = 'purple')
#Selanjutnya kita memanggil metode untuk membuat histogram
#variabel hist sebagai objek dari diagram batang tersebut
#Kita juga dapat menentukan warna pada histogram dengan menambahan parameter color
#Variabel hist berisi sebuah list yang berisi bar chart yang telah dibuat.
pola = ['/','/','/','*','*','*','*']
#Kita juga dapat membuat sebuah list 'pola' yang berisi simbol-simbol yang digunakan sebagai pola pada bar chart
for bar in hist :
#digunakan loop 'for' untuk mengakses setiap bar pada histogram variabel hist
    bar.set_hatch(pola.pop(0))
#pada setiap bar, diatur pola dengan menggunakan set_hatch dan pola.pop(0)
#pola.pop(0) digunakan untuk mengambil dan menghapus elemen pertama dari list pola sehingga setiap bar akan memiliki pola yang berbeda
plt.show()
#kode diatas dipanggil untuk menampilkan histogram yang telah dibuat
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# membaca file data iris
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                 names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])

# menampilkan lima data pertama dari dataframe
print(df.head(9))

# menghitung nilai rata-rata setiap variabel untuk setiap jenis bunga iris
df_mean = df.groupby(['class']).mean()
print(df_mean)

# membuat plot scatter untuk melihat persebaran bunga iris berdasarkan panjang dan lebar kelopak bunga
plt.scatter(df[df['class']=='Iris-setosa']['sepal length'], df[df['class']=='Iris-setosa']['sepal width'], 
            color='green', label='Iris-setosa')
plt.scatter(df[df['class']=='Iris-versicolor']['sepal length'], df[df['class']=='Iris-versicolor']['sepal width'], 
            color='red', label='Iris-versicolor')
plt.scatter(df[df['class']=='Iris-virginica']['sepal length'], df[df['class']=='Iris-virginica']['sepal width'], 
            color='blue', label='Iris-virginica')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Scatter plot of Iris dataset')
plt.legend()
plt.show()
```
